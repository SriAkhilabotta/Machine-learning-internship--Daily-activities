# Z-Score

```python
import pandas as pd
data=pd.DataFrame({
    "study_hours":[1,2,3,4,5,6,7,8,20]
})
```

```python
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
z_scores=np.abs(stats.zscore(data["study_hours"]))
outliers=data[z_scores>2]
outliers
```

```python
plt.scatter(range(len(data)),data["study_hours"])
plt.xlabel("Index")
plt.ylabel("Study hours")
plt.title("scatter plot showing outlier")
plt.show()
```

```python
plt.hist(data["study_hours"],bins=40)
plt.xlabel("study hours")
plt.ylabel("frequency")
plt.title("his plot showing outlier")
plt.show()
```
