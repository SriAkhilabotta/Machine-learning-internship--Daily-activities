# Equal-width binnning


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "Age":[18,22,25,30,35,40,45,50]
})
print(data)
```

```python
data["Age_bin"]=pd.cut(data["Age"],bins=4)
data
```

```python
plt.hist(data["Age"],bins=3)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Width Binning using pd.cut()")
plt.show()
```

```python
plt.scatter(range(len(data)),data["Age"])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Scatter Plot Showing Binning")
plt.show()
```

```python
data["Age_frequency_bin"] = pd.qcut(data["Age"], q=4)
data
```
