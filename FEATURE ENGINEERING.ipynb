# FEATURE ENGINEERING:
- Feature engineering is the process of converting raw data into meaningful numerical features so that a ml model can understand patterns and make better prediction.

1.Feature exploration:
-----------------------
- The process of anayzing and visualizing features to uncover patterns,distributions,coreelations and potenttial issues in the dataset before applying ml.

2.Feature creation:
--------------------------
- The process of generating new meaningful features from existing data using domain knowledge or mathematical operations to help the model learn better patterns.

3.Feature Encoding:
----------------------------
- The process of converting categorical or non numeric features into numeric from so machine learning models can process them effectively

4.Feature Scaling:
-------------------------
- The process of transforming numerical features to a standard range so that no feature dominates the learning process due to its scale

5.Feature Selection:
---------------------------
- The process of identifying and keeping only the most relevant features while removing irrelavant or redundant ones to improve model perfromance and reduce overfitting.

-> Common Types of Feature creation:
------------------------------------------------------

* Domain- based features(real-world Logic)
* mathematical combinations(sum,average,ratio)
* Interaction features( combined effect of features)
* polynomial features(non linear relationships)

-> Examples:
* Total_marks= Math+Science
* efficiency= total_marks/Study_hours

-> Impact on model:
----------------------------
- Improves accuracy
- Helps models capture complex patterns
- Makes simple models perfom better

```python
import pandas as pd
df=pd.DataFrame({
    "Study_Hours":[2,4,6,8,10],
    "Attendance":[60,70,80,90,95],
    "Maths":[40,55,65,75,80],
    "science":[42,58,68,78,88]
})
df
```

# Domain-based feature

```python
df["Total_marks"]=df["Maths"]+ df["science"]
df
```

# Mathematical Feature

```python
df["marks_perhour"]=df["Total_marks"]/df["Study_Hours"]
df
```

# Interaction Feature

```python
df["Study_attedendance_interaction"]=df["Study_Hours"]*df["Attendance"]
df
```

# Polynomial Feature

```python
df["Study_Hours_squared"]=df["Study_Hours"]**2
df
```
